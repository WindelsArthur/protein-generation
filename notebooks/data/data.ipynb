```python
import pandas as pd

# Votre code existant
protein_data = pd.read_csv('/home/arthur/projets/protein-generation/data/train.tsv', sep=',')

# Calculer les longueurs des séquences
protein_data['sequence_length'] = protein_data['sequence'].str.len()

# DÉFINIR VOS PROPRES VALEURS L_min et L_max
L_min = 380  # Changez cette valeur
L_max = 500  # Changez cette valeur

print(f"Filtrage entre {L_min} et {L_max} caractères")

# Filtrer les données selon vos critères
filtered_data = protein_data[(protein_data['sequence_length'] >= L_min) & 
                            (protein_data['sequence_length'] <= L_max)]

print(f"Nombre de séquences dans la plage [{L_min}, {L_max}] : {len(filtered_data)}")
print(f"Nombre total de séquences : {len(protein_data)}")

# Compter le nombre de séquences pour chaque longueur dans la plage filtrée
longueurs_count = filtered_data['sequence_length'].value_counts()

# Trier par nombre de séquences (décroissant) et afficher les 30 premières
top_30_longueurs = longueurs_count.head(100)

print(f"\nTop 30 des longueurs avec le plus de séquences (entre {L_min} et {L_max}) :")
print("Longueur | Nombre de séquences")
print("-" * 30)
for longueur, count in top_30_longueurs.items():
    print(f"{longueur:8d} | {count:6d}")
```
